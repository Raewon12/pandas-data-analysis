```python
url = "https://raw.githubusercontent.com/pia222sk20/python/refs/heads/main/pandas-data-analysis/part3/data/train.csv"
```

```python
import pandas as pd
df = pd.read_csv(url)
df.head()
```

```python
df.hour.value_counts()
```

```python
df[df.hour== 20]
```

```python
df.info()
```

```python
# 시간에 따른 자전거 대여량 변화 추이
# 시간별 자전거 대여량

# count 열의 합계를 시간별로 그룹화
hour_sum_df = df.groupby('hour')['count'].sum()

# 그래프 그리기
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
hour_sum_df.plot(kind='line', marker='o')
plt.title('시간대별 자전거 대여량')
plt.xlabel('시간')
plt.ylabel('대여량')
plt.grid(True)
plt.show()

# 시간대별 대여량 데이터 출력
print("\n시간대별 대여량:")
print(hour_sum_df)
```
